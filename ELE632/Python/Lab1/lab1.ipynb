```python
# I. Delta[n-3]
import numpy as np 
import matplotlib.pyplot as plt

def u(a, n):
    unit = np.array([])
    for sample in n:
        if sample<a:
            unit = np.append(unit,0)
        else:
            unit = np.append(unit,1)
    return(unit)



n = np. arange(-3,7)

plt.figure()
plt.stem(n, u(3,n)-u(4,n))
plt.show()
```

```python
# II. u[n+1]

plt.figure()
plt.stem(n, u(1,n))
plt.show()
```

```python
# III. x[n] = cos(pi*n/5)u[n]

x = np.cos(np.pi*n/5)

plt.figure()
plt.stem(n,x*u(0,n))
plt.show()
```

```python
# IV. x1[n] = x[n-3]

n = np.arange(0,12)
x = np.cos(np.pi*(n-3)/5)

plt.figure()
plt.stem(n, x*u(3,n))
plt.show()
```

```python
# V. x2[n] = x[-n]

n = np.arange(-11,3)
x = np.cos(np.pi*(-n)/5)

plt.figure()
plt.stem(n, x*u(0,-n))
plt.show()
```

x1[n] = x[n-3] is advance by 3 
x2[n] = x[-n] is reflection in the y axix

#### D. Energy and power of a discrete signal 

##### 1)

```python
def energy(x):
    sum = 0
    for i in x:
        term = np.absolute(x)**2
        sum += term
    return sum

    
def power(x, N):
    sum = 0
    for i in x:
        term = np.absolute(x)**2/(2*N+1)
        sum += term
    return sum
```
